# Use this notebook for your 3 choices of APIs

```python
import requests
import json
import pandas as pd
pd.set_option('display.max_columns', None)
```

* Extracting data form Fruityvice.com.
* Project idea: predict the number of calories of a fruit based on other informations about it

```python
response = requests.get('https://www.fruityvice.com/api/fruit/all')
response.content              #json structure

results = response.json()     #convert json into python language
fruit_df = pd.json_normalize(results)
fruit_df
```

* Extracting data from fbi wanted api
* Project idea: descriptive statistics about wanted criminals (where they commit crimes, their appearance, etc.)

```python
response = requests.get('https://api.fbi.gov/wanted/v1/list')
response.content              #json structure

results = response.json()     #convert json into python language
results

wanted_df = pd.json_normalize(results["items"])
wanted_df
```

* Extracting data from GGBIF species api
* Project idea: discover and access information about species and higher taxa, interpreting names, looking up the identifiers and scientific names

```python
response = requests.get('https://api.gbif.org/v1/species')
response.content              #json structure

results = response.json()     #convert json into python language
results

species_df = pd.json_normalize(results["results"])
species_df
```
